# Limpiar datos

**Extracción de archivos .dat desde catálogos OGLE**

Este notebook descomprime automáticamente los archivos descargados de OGLE y los extrae de manera automática. El archivo de salida es un archivo .csv con la información extraída.


```python
# Librerías
import os
import shutil
import tarfile
import zipfile
```

```python
# Carpeta donde están los archivos comprimidos
LMC = "BD/OGLEIV/RR_Lyrae/RR_Lyrae_LMC (La gran nube de Magallanes)"
SMC = "BD/OGLEIV/RR_Lyrae/RR_Lyrae_SMC (Pequeña nube de Magallanes)"
BLG = "BD/OGLEIV/RR_Lyrae/RR_Lyrae_BLG (Bulbo galáctico)"
input_dir = input(" Ingresar ruta de la carpeta\n 1. LMC\n 2. SMC\n 3. BLG\n")

# Selecionar carpeta según opción
if input_dir == "1":
    input_dir = LMC
elif input_dir == "2":
    input_dir = SMC
elif input_dir == "3":
    input_dir = BLG
else:
    print(" Opción no válida. Usando LMC por defecto.")
    input_dir = LMC


# Carpeta donde se guardarán los .dat
output_dir = "BD/OGLEIV/RR_Lyrae/limpios/SMC_limpios"

# Crear carpeta de salida si no existe
os.makedirs(output_dir, exist_ok=True)

def extraer_archivos(archive_path, output_dir):
  """
  Extrae archivos .dat de un archivo compromido (tar.gz o zip) y los guarda en la carpeta de destino.
  """
  try:
    # Caso .tar.gz o .tar
    if archive_path.endswith((",tar.gz", "tar")):
      with tarfile.open(archive_path, "r:*") as tar:
        for member in tar.getmembers():
          if member.name.endswith(".dat"):
            tar.extract(member, output_dir)
  
    # Caso .zip
    elif archive_path.endswith(".zip"):
      with zipfile.Zipfile(archive_path, "r") as zf:
        for file in zf.namelist():
          if file.endswith(".dat"):
            zf.extract(file, output_dir)

    print(f"Procesado: {os.path.basename(archive_path)}")
  
  except Exception as e:
    print(f"Error con {archive_path}: {e}")

# Recorre todos los archivos comprimidos en la carpeta input_dir
for fname in os.listdir(input_dir):
  fpath = os.path.join(input_dir, fname)
  if os.path.isfile(fpath) and (fname.endswith((".tar,gz", ".tar", ".zip"))):
    extraer_archivos(fpath, output_dir)

print("\n Extracción completa. Archivos .dat guardados en:", output_dir)

```
